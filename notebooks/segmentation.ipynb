## Imports



```python
import os

import numpy as np

from skimage.io import imread
import matplotlib.pyplot as plt

import napari
from napari.utils import nbscreenshot

from napari_bio_sample_data._sample_data import nuclei_data
```

## Data



```python
dir = os.path.dirname("/Users/kharrington/mambaforge/envs/imaging-demo-kit/lib/python3.1/site-packages/napari_bio_sample_data/")

# Load data taken from: https://github.com/chili-chiu/napari-bio-sample-data/blob/8d8177e7a829754010d87826a753ff8856e9f87e/src/napari_bio_sample_data/_sample_data.py#L29-L34

img = imread(os.path.join(dir,'./sample_images/nuclei.tif'))
label = imread(os.path.join(dir,'./sample_images/nuclei_label.tif'))
surface_read = np.load(os.path.join(dir,'./sample_images/nuclei_surface.npz'))
surface = tuple(surface_read.values())
```

## Visualizing



```python
viewer = napari.Viewer()

viewer.add_image(img, name="nuclei")
viewer.add_labels(label, name="nuclei label")
viewer.add_surface(surface, name="nuclei surface")

nbscreenshot(viewer)
```
